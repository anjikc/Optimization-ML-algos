## Hierarchical Clustering 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Create a blob of 200 data points
data_set=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=50)
data_set
```

```python
points=data_set[0]
import scipy.cluster.hierarchy as sch 
from sklearn.cluster import AgglomerativeClustering 
```

```python
print(points)
```

```python
# Creating a dendogram 
dendrogram=sch.dendrogram(sch.linkage(points,method='ward'))
```

```python
## Creating a scatter plot of the data points
plt.scatter(data_set[0][:,0], data_set[0][:,1])
```

```python
## Perform hierarchical clustering on the dataset
hier_clust=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
```

```python
y_hier_clust=hier_clust.fit_predict(points)
```

```python
print(y_hier_clust)
```

```python
plt.scatter(points[y_hier_clust==0,0],points[y_hier_clust==0,1],s=100,c='cyan')
plt.scatter(points[y_hier_clust==1,0],points[y_hier_clust==1,1],s=100,c='yellow')
plt.scatter(points[y_hier_clust==2,0],points[y_hier_clust==2,1],s=100,c='red')
plt.scatter(points[y_hier_clust==3,0],points[y_hier_clust==3,1],s=100,c='green')
```
